```python
from google.colab import drive
import pandas as pd

drive.mount('/content/drive')
```

```python
file_path_base = '/content/drive/MyDrive'
```

```python
df_acm = pd.read_csv(f'{file_path_base}/Relevant(ACM).csv')
df_ieee = pd.read_csv(f'{file_path_base}/Relevant(IEEE).csv')
df_sd = pd.read_csv(f'{file_path_base}/Relevant(Science_direct).csv')
```

```python
df = pd.concat([df_acm, df_ieee, df_sd])
df.reset_index()
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
columns = ['Title', 'Author', 'Abstract Note', 'Publication Year', 'Item Type',
       'Publication Title', 'ISBN', 'ISSN', 'DOI', 'Url',
       'Date', 'Pages', 'Volume', 'Publisher','Library Catalog']


result = df[columns]

result['Item Type'] = result['Item Type'].replace({'conferencePaper': 'Conference Paper', 'journalArticle': 'Journal Article'})

result.head()
```

```python
result['Library Catalog'].loc[~df['Library Catalog'].isin(['ACM Digital Library', 'IEEE Xplore'])] = 'Science Direct'
result['Library Catalog'].value_counts()
```

```python
result.head()
```

```python
df_acm = result[df['Library Catalog'].isin(['ACM Digital Library'])]
df_ieee = result[df['Library Catalog'].isin(['IEEE Xplore'])]
df_sd = result[~df['Library Catalog'].isin(['ACM Digital Library', 'IEEE Xplore'])]
```

```python
print(f"{df_acm.shape=} \n{df_ieee.shape=} \n{df_sd.shape=}\nTotal: {df_acm.shape[0]+df_ieee.shape[0]+df_sd.shape[0]}")
```

```python
df_acm.to_csv(f"{file_path_base}/acm.csv")
df_ieee.to_csv(f"{file_path_base}/ieee.csv")
df_sd.to_csv(f"{file_path_base}/sd.csv")
```
